# Reading Data

Using pandas.read_csv() function to read the CSV file. 
In the brackets, put the file path along with a quotation mark so that Pandas will read the file into a data frame from that address. 
The file path can be either an URL or your local file address.

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("sales.csv")

# Display the first 5 rows
df.head()
```

```python
# Display the last 5 rows
df.tail(5)
```

# Quick Visualiztion of Sales Data
We can show a quick visualiztion of Price Distribution using the 'Price of Each' column of Product in the sales data.

```python
# Quick visualization
import matplotlib.pyplot as plt
%matplotlib inline
df['Price Each'].plot(kind='hist', bins=20, title='Price Distribution')
plt.show()
```
